```python
import pandas as pd
import sqlite3
import os
```

```python
print("Current Directory:")
print(os.getcwd())
```

```python
print(os.listdir("../data/raw"))
```

```python
df = pd.read_excel(
    "../data/raw/credit_default_raw.xls",
    header=1
)

df.head()
```

```python
print("Dataset Shape:")
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.columns.tolist())
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
conn = sqlite3.connect(
    "../database/credit_risk.db"
)

print("Database Connected")
```

```python
df.to_sql(
    "credit_data",
    conn,
    if_exists="replace",
    index=False
)

print("Data Saved To SQLite")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*)
FROM credit_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM credit_data
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(AGE)
AS average_age
FROM credit_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT EDUCATION,
COUNT(*) AS total
FROM credit_data
GROUP BY EDUCATION;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM credit_data
WHERE LIMIT_BAL > 300000
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
conn.close()

print("Connection Closed")
```
